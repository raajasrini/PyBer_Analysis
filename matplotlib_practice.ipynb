```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim("Jan", "June")
plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.figure(figsize=(20, 10))
#plt.subplots(figsize=(20, 10))
```

```python
plt.plot(x_axis, y_axis, label='Boston',c='green')
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x_axis, y_axis, label='Boston',c='green',marker="D")
plt.grid()
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim("Jan", "June")
plt.xlim(0,11)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*5)
    
#ax.scatter(x_axis, y_axis, s=y_axis_larger,label='Boston',c='skyblue',marker='o',edgecolor='black',linewidth=2)
plt.legend()
plt.title("Pyber Fare by Month")
ax.scatter(y_axis, x_axis, s=y_axis_larger,label='Boston',c='skyblue',marker='o',edgecolor='black',linewidth=2)
plt.gca().invert_yaxis()
plt.xlabel('Fare ($)')
plt.ylabel('Date ')
plt.xlim(0,50)
plt.show()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#ax.pie(y_axis,labels=x_axis)
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow = True,
    startangle=90,
    counterclock = False)

plt.show()
```
